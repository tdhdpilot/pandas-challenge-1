## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a new column for client 33615 and 66037 that calculates the subtotal for each line using the unit_price and the qty. show unit_price, qty, and subtotal sum for the first 5 rows of the dataframe
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['subtotal'] = df['unit_price'] * df['qty']

# Filter the dataframe for client 33615 and 66037
filtered_df = df[df['client_id'].isin([33615, 66037])]
total_spent_by_client = filtered_df.groupby('client_id')['subtotal'].sum().reset_index(name='total_spent')

client_spent_most = total_spent_by_client.loc[total_spent_by_client['total_spent'].idxmax()]
print("Total spent by each client:")
print(total_spent_by_client.to_string(formatters={'total_spent':'${:,.2f}'.format}))
print(f"\nClient ID that spent the most: {client_spent_most['client_id']} with a total of ${client_spent_most['total_spent']:,.2f}")
df.head()                                  


```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

def calculate_shipping_price(row):
    shipping_weight = row['unit_weight'] * row['qty']
    if shipping_weight > 50:
        return shipping_weight * 7
    else:
        return shipping_weight * 10
df['shipping_price'] = df.apply(calculate_shipping_price, axis=1)
df.head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
df['sales_tax_rate'] = sales_tax_rate = 0.0925
df['total_price'] = (df['subtotal'] + df['shipping_price']) * (1 + sales_tax_rate)
df['total_price'] = df['total_price'].apply(lambda x: (x)). round(2)
#df[‘total_price’] = ((df[‘subtotal’] + df[‘shipping_price’]) * (1 + df[‘sales_tax_rate’])).round(2)
df.head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df['cost'] = df['cost'].apply(lambda x: (x))
df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price

df['profit'] = df['unit_price'] - df['unit_cost']
# Format the ‘profit’ column
df['profit'] = df['profit'].apply(lambda x: (x))
df.head()
```
